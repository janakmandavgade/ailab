```python
import numpy as np

def perceptron_learning(gate, learning_rate=0.1, max_iterations=1000):
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    y = np.array(gate)

    np.random.seed(42)
    w = np.random.uniform(1, 10, size=(2,))
    b = np.random.uniform(1, 10)
    
    for iteration in range(max_iterations):
        errors = 0
        
        for i in range(len(X)):
            x = X[i]
            target = y[i]
            output = np.dot(w, x) + b

            if output >= 0:
                y_pred = 1
            else:
                y_pred = 0
                
            delta = target - y_pred
            if delta != 0:
                errors += 1
                w += learning_rate * delta * x
                b += learning_rate * delta
                
        if errors == 0:
            break
            
    return w, b, iteration + 1

AND_GATE = [0, 0, 0, 1]
OR_GATE = [0, 1, 1, 1]
NAND_GATE = [1, 1, 1, 0]
NOR_GATE = [1, 0, 0, 0]

w_and, b_and, iter_and = perceptron_learning(AND_GATE)
print("AND gate completed in {} iterations.".format(iter_and))

w_or, b_or, iter_or = perceptron_learning(OR_GATE)
print("OR gate completed in {} iterations.".format(iter_or))

w_nand, b_nand, iter_nand = perceptron_learning(NAND_GATE)
print("NAND gate completed in {} iterations.".format(iter_nand))

w_nor, b_nor, iter_nor = perceptron_learning(NOR_GATE)
print("NOR gate training in {} iterations.".format(iter_nor))
```
